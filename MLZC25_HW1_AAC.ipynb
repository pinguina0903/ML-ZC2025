```python
df_asia = data[data.origin == 'Asia']
df_asia = df_asia[['vehicle_weight','model_year']]
df_asia
```

```python
df_q7 = df_asia.iloc[ : 7]
```

```python
df_q7
```

```python
X = df_q7.values
XTX = X.T.dot(X)

XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = (XTX_inv @ X.T) @ y
w
```

```python
sum(w)
```
